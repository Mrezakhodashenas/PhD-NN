### Making one customer per row for Accounts

```python
print('shape df_accounts_train:' , df_accounts_train.shape)
# create a new column with a value of 1
df_accounts_train['Has_Account'] = 1

# pivot the data by customer
df_accounts_train_pivoted = pd.pivot_table(df_accounts_train, values='Has_Account', index='Customer', columns='Account', fill_value=0)

# reset the index to make Customer a column again
df_accounts_train_pivoted = df_accounts_train_pivoted.reset_index()

# check the pivoted dataframe
print('df_accounts_train_pivoted shape:',df_accounts_train_pivoted.shape)

df_accounts_train_pivoted.head()
```

## Making the DataFrame (1 customer/row) by merging all three datasets:

```python
df_merged = df_customers_train.merge(df_transactions_train, on='Customer', how='left').merge(df_accounts_train_pivoted, on='Customer', how='left')
print(df_merged.shape)
df_merged.head(20)


# summarize the transaction and account data by customer
df_summary = df_merged.groupby('Customer').agg({
    'Churn': 'first',
    'Card': 'first',
    'Start_Date': 'first',
    'Customer_Service_Calls': 'first',
    'Credit_Limit': 'first',
    'Total_Revolving_Balance': 'first',
    'Average_Open_To_Buy': 'first',
    'Average_Utilisation_Ratio': 'first',
    'Age': 'first',
    'Gender': 'first',
    'Education': 'first',
    'Marital_Status': 'first',
    'Dependents': 'first',
    'Income': 'first',
    'Credit Card': 'first',
    'Current': 'first',
    'Deposit': 'first',
    'Investment': 'first',
    'Joint': 'first',
    'Loan': 'first',
    'Mortgage': 'first',
    'On Demand Deposit': 'first',
    'Date': ['min', 'max', 'count'],
    'Amount': ['sum', 'mean']
})

# flatten the multi-index column names
df_summary.columns = ['_'.join(col) for col in df_summary.columns.values]

# reset the index to make Customer a column again
df_summary = df_summary.reset_index()

# check the summarized dataframe
print(df_summary.shape)
df_summary.columns = [col.replace('_first', '') for col in df_summary.columns]
df_summary.columns = [col.replace('Date_min', 'First_trans') for col in df_summary.columns]
df_summary.columns = [col.replace('Date_max', 'Last_trans') for col in df_summary.columns]

df=df_summary
```

### Creating Box Plot (with and without Outliers)

```python
# ----------------- handle outliers using the IQR method
Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1
df_num_out = df_num[~((df_num < (Q1 - 1.5 * IQR)) |(df_num > (Q3 + 1.5 * IQR))).any(axis=1)]
# -----------------

# print the number of rows with and without outliers
print(f"Number of rows with outliers: {len(df_num)}")
print(f"Number of rows without outliers: {len(df_num_out)}")
```

```python
# define the numerical columns
numerical_cols = ['Customer_Service_Calls', 'Credit_Limit', 'Average_Open_To_Buy', 'Amount_sum', 'Amount_mean']


# create a figure with two columns and as many rows as numerical columns
fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=2, figsize=(10, 12))

# iterate over the numerical columns and plot the box plots
for i, col in enumerate(numerical_cols):
    q1 = df_num[col].quantile(0.25)
    q2 = df_num[col].quantile(0.5)
    q3 = df_num[col].quantile(0.75)
    
    # plot the box plot
    sns.boxplot(x=df_num[col], ax=axes[i, 0])
    axes[i, 0].set_title(f"{col}\nMedian: {q2}\nQ1: {q1}\nQ3: {q3}",  fontsize=7)
   


    # plot the box plot after handling ------------- outliers-------------
    q1 = df_num_out[col].quantile(0.25)
    q2 = df_num_out[col].quantile(0.5)
    q3 = df_num_out[col].quantile(0.75)

    sns.boxplot(x=df_num_out[col], ax=axes[i, 1])
    axes[i, 1].set_title(f"{col}(After handling Outliers) \n Median: {q2}\nQ1: {q1}\nQ3: {q3}",  fontsize=7)

    # calculate and display the mean for each numerical column
    mean_val = df_num[col].mean()
    axes[i, 0].axvline(mean_val, color='r', linestyle='--')
    axes[i, 1].axvline(mean_val, color='r', linestyle='--')
    axes[i, 0].text(mean_val, 0.95, f"Avg = {mean_val:.2f}", transform=axes[i, 0].get_xaxis_transform(),fontsize=7)
    axes[i, 1].text(mean_val, 0.95, f"Avg = {mean_val:.2f}", transform=axes[i, 1].get_xaxis_transform(),fontsize=7)

# adjust the spacing between subplots
fig.subplots_adjust(hspace=1.5, wspace=0.2)
# show the figure
plt.show()
plt.savefig("output/Box Plot (with and without Outliers).png")

```

1.   The above figure shows that outliers on the higher end of '*Customer_Service_Calls*', '*Credit_Limit*',  '*Average_Open_To_Buy*', '*Amount_sum*', and '*Amount_mean*'. 
2.   Except *Customer_Service_Calls*, which its distribution seems relatively normal, for the rest of them the distribution is skewed to the right.

### Customer Churn Rate Over Time

```python
# Group the data by month and Churn, and calculate the churn rate
churn_by_month = df_new.groupby([pd.Grouper(key='Start_Date', freq='M'), 'Churn'])['Customer'].count().unstack()
churn_rate_by_month = churn_by_month['Yes'] / churn_by_month.sum(axis=1)

# Find the top 3 months with maximum churn rate
top3_months = churn_rate_by_month.argsort()[-3:][::-1]
top_3_min_churn = churn_rate_by_month.sort_values().head(3)

# Plot the churn rate over time, highlighting the top 3 months
plt.plot(churn_rate_by_month.index, churn_rate_by_month.values, color='blue')
plt.scatter(churn_rate_by_month.index[top3_months], churn_rate_by_month.values[top3_months], color='red', marker='o', s=100)
plt.scatter(top_3_min_churn.index, churn_rate_by_month.loc[top_3_min_churn.index], color='green', marker='o', s=100)
plt.xlabel('Month')
plt.ylabel('Churn rate')
plt.title('Customer Churn Rate Over Time')
plt.show()
plt.savefig("output/Customer Churn Rate Over Time.png")
```



*   The above figure shows that the Churn rate went to maximum rates in the **last quarter of 2018**, and also in the **2nd half of 2021**.


*   Also, the minimum Churn rate happened a little earlier than the first max, roughly in the **3rd quarter of 2018**. Moreover, before and after the maximum Churn rate in the **2nd half of 2021**, there were two minimum Churn rate. So, the **2nd half of 2021** was a fluctuation in the rate of Churned customer, which might be interesting to investigate why or what happened.


*   It is worth mentiong that the fluctuations in the Churn rate,happened in the 2nd half of the two years (2018, 2021)




### Durations of accounts based on Last and First Trans and Start date by Churn target

```python
# Calculate the duration between each date column and the current date

# df.Start_Date_Duration = ( df_new.Last_trans - df_new.Start_Date).dt.days
# df.Trans_Duration = ( df_new.Last_trans - df_new.First_trans).dt.days
df['Start_Date_Duration'] = (df['Last_trans'] - df['Start_Date']).dt.days
df['Trans_Duration'] = (df['Last_trans'] - df['First_trans']).dt.days

# Plot the distribution of each duration by churn status
fig, axs = plt.subplots(nrows=2, figsize=(8, 8))
axs[0].hist(df.loc[df["Churn"] == "Yes", "Start_Date_Duration"], alpha=0.5, label="Churned")
axs[0].hist(df.loc[df["Churn"] == "No", "Start_Date_Duration"], alpha=0.5, label="Not Churned")
axs[0].set_xlabel("Duration from Start date (days)")
axs[0].set_ylabel("Count")
axs[0].set_title("Duration btw Last Trans & account start date (days)")
axs[0].legend()

axs[1].hist(df.loc[df["Churn"] == "Yes", "Trans_Duration"], alpha=0.5, label="Churned")
axs[1].hist(df.loc[df["Churn"] == "No", "Trans_Duration"], alpha=0.5, label="Not Churned")
axs[1].set_xlabel("Transaction Duration (days)")
axs[1].set_ylabel("Count")
axs[1].set_title("Duration btw Last & First trasnactions (days)")
axs[1].legend()

plt.tight_layout()
plt.show()
plt.savefig("output/Durations of accounts based on Last and First Trans and Start date by Churn target.png")
```


*   The above plot shows two histograms that represent the distribution of the duration between the start date and the last transaction date and the duration between the first and last transaction date for two groups: customers who churned and customers who did not churn. 

*   The first histogram shows that most customers who did not churn made their last transaction within the first 1000 days of their account opening, while the churned customers have a more uniform distribution across the same range. 
*   The second histogram shows a heavily skewed distribution to the left, and implies that most customers churn after around 360 days of their first transaction (Obviously they don't churn in the first days of recieving the service but eventually they churn after around one year) 

*   There is a gap metween 300 and 320 days after the first transaction
*   Again, it is important to know that all the above analysis is just based on the number of churns not the percentage of them.
